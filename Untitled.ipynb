```python
# Load Excel data from Table 5.6, sheet 'Annual'
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

ipi = pd.read_excel('hackathon_ipi3.xlsx')
ipi
```

```python
time1 = ipi.loc[:,'TIME_Y']
time2 = ipi.loc[:,'TIME_M']
time3 = ipi.loc[:,'TIME_Q']
print(time1[0:12])
print(time2[0:12])
print(time3[0:12])
```

```python
value = ipi.loc[:,'Value']
value2 = ipi.loc[:,'Value.1']
value3 = ipi.loc[:,'Value.2']
print(value[0:12])

print(value2[0:12])

print(value3[0:12])
```

```python
plt.figure(1)
plt.plot(time3,value3)
plt.legend(['YEAR_FREQUENCY','Production'])
plt.xlabel('Time_Quarters')
plt.ylabel('Production')
plt.savefig('myplot.png')
```

```python
plt.figure(2)

plt.plot(time1,value2)
plt.legend(['YEAR_FREQUENCY','Production'])
plt.xlabel('Time_Years')
plt.ylabel('Production')
plt.savefig('myplot.png')
```

Both graphs above exhibit a clear reduction in production accross the economy. Looking at the monthly data, there is a close relationship between the economic crash of 2007-2009 to the present ongoing crash. The first graph (quarterly) suggests an even greater hit on the production market. However, considerding this, the general rise in the stock market is concerning as a rise in currency is not an appropriate reflection of the decrease in production since the two contradict eachother.

```python
# Load Excel data from Table 5.6, sheet 'Annual'
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

commodities_df = pd.read_excel('hackathon_commodities.xlsx')
commodities_df.head()

# Isolate Gold from the file
gold = commodities_df.loc[:,'Gold']
aluminium = commodities_df.loc[:,'Aluminium']
brent = commodities_df.loc[:,'Brent']
date = commodities_df.loc[:,'Date']
print(gold[0:12])

print(aluminium[0:12])

print(brent[0:12])

print(date[0:12])
```

```python
plt.figure(3)

plt.plot(date,brent)
plt.plot(date,gold)
plt.plot(date,aluminium)
plt.legend(['Brent','Gold','Aluminium'])
plt.xlabel('Time_Years')
plt.ylabel('Production')
plt.savefig('myplot.png')
```
